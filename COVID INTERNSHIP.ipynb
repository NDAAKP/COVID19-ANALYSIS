```python
covid_df.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],inplace=True,axis=1)
```

```python
covid_df.head()
```

```python
covid_df.tail()
```

```python
covid_df['Date']=pd.to_datetime(covid_df['Date'], format='%Y-%m-%d')
```

```python
covid_df
```

```python
covid_df['Active Cases']=covid_df['Confirmed']-covid_df['Cured']-covid_df['Deaths']
```

```python
covid_df
```

```python
statewise=pd.pivot_table(covid_df,values=["Confirmed","Deaths","Cured","Active Cases"],index="State/UnionTerritory",aggfunc=max)
```

```python
statewise
```

```python
statewise["Recovery Rate"]=(statewise["Cured"]/statewise["Confirmed"])*100
```

```python
statewise["Mortality Rate"]=(statewise["Deaths"]/statewise["Confirmed"])*100
```

```python
statewise
```

```python
statewise_confirmed=statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise_confirmed
```

```python
statewise_confirmed.style.background_gradient(cmap='cubehelix')
```

```python
top_10_active_cases=covid_df.groupby(by="State/UnionTerritory").max()[['Active Cases','Date']].sort_values(by=['Active Cases'],ascending=False).reset_index()
```

```python
import matplotlib.pyplot as plt
top_10_active_cases=covid_df.groupby(by="State/UnionTerritory").max()[['Active Cases','Date']].sort_values(by=['Active Cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 Most Active Cases in India", size=25)
ax=sns.barplot(data=top_10_active_cases.iloc[:10], y="Active Cases", x="State/UnionTerritory", linewidth=3, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Active Cases")
plt.show()
```

```python
top_10_death_cases=covid_df.groupby(by="State/UnionTerritory").max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 Most Deaths in India", size=25)
ax=sns.barplot(data=top_10_death_cases.iloc[:12], y="Deaths", x="State/UnionTerritory", linewidth=3, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Deaths")
plt.show()
```

```python
top_10_confirmed_cases=covid_df.groupby(by="State/UnionTerritory").max()[['Confirmed','Date']].sort_values(by=['Confirmed'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 Most Confirmed Cases in India", size=25)
ax=sns.barplot(data=top_10_confirmed_cases.iloc[:12], y="Confirmed", x="State/UnionTerritory", linewidth=3, edgecolor='black')
plt.xlabel("States")
plt.ylabel("Confirmed")
plt.show()
```

```python
fig=plt.figure(figsize=(16,9))
ax=sns.lineplot(data=covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])] ,x='Date',y='Active Cases',hue='State/UnionTerritory')
ax.set_title("Top 5 State's Active Cases Growth Over Time", size=20)
```

```python
vaccine=vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
vaccine
```

```python
max_vac=vaccine.groupby('State')['Total'].sum().to_frame("Total")
max_vac=max_vac.sort_values("Total",ascending=False)[:5]
max_vac
```

```python
min_vac=vaccine.groupby('State')['Total'].sum().to_frame("Total")
min_vac=min_vac.sort_values("Total",ascending=True)[:5]
min_vac
```

```python
fig=plt.figure(figsize=(20,20))
plt.title("States with most Vaccinations",size=20)
x=sns.barplot(data=max_vac.iloc[:10],y=max_vac.Total,x=max_vac.index,linewidth=3,edgecolor='black')
plt.xlabel("States",size=20)
plt.ylabel("No. of Vaccinations",size=20)
plt.show()
```

```python
fig=plt.figure(figsize=(20,20))
plt.title("States with least Vaccinations",size=20)
x=sns.barplot(data=min_vac.iloc[:10],y=min_vac.Total,x=min_vac.index,linewidth=3,edgecolor='black')
plt.xlabel("States",size=20)
plt.ylabel("No. of Vaccinations",size=20)
plt.show()
```
